## What's a sine function? 

Sine (or Sinus) is actually a trigonometric function of an angel. I think this is enough, because there's more and better information on this topic on [Wikipedia](https://en.wikipedia.org/wiki/Sine). 

## Why we need sine function? 

Because a sine function is actually a great help for us in making a _Sine Wave_. We'll study how sound wave (or pure sonic signals) work later. But first, let's just graph a sine function using python. 

## Importing Packages

````python
import numpy as np
import matplotlib.pyplot as plt 
````

Now, we have two packages we need. Then, we have to define a value for our function. It's called an _independent value_ and X axis, is for this value/variable. Also, we have define a function for our initial value. It's called _dependent value_ and Y axis is for that. 

````python
x = np.arange(0, 10, 0.1)
# It starts on 0, ends on 10. The spacing is 0.1. 
y = np.sin(x)
````

Now we initialized our values. We just need to plot that to see how a sine wave looks like. 

````python
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Sin(X)')
````

Now what we need is calling a function to show us our plot (If you use terminal for testing these codes). 

````python
plt.show()
````
